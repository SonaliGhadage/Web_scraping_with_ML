### **TASK 1** 

**Revenue Prediction Model**

```python
all_data = pd.read_excel("/content/drive/My Drive/Colab Notebooks/data.xlsx", sheet_name=None)


df1 = all_data['Products_ActBalance']
df2 = pd.merge(all_data['Inflow_Outflow'], all_data['Sales_Revenues'], on=['Client'])

final_dataframe = pd.merge(df1, df2, on=['Client'])
print(final_dataframe.shape)
final_dataframe.head()
```

```python
# combining all sales and revenues

final_dataframe['Sales'] = final_dataframe['Sale_MF'] + final_dataframe['Sale_CC'] + final_dataframe['Sale_CL']
final_dataframe['Revenues'] = final_dataframe['Revenue_MF'] + final_dataframe['Revenue_CC'] + final_dataframe['Revenue_CL']

# modified dataframe with new "sales" and "revenues"

new_dataframe = final_dataframe.drop(['Sale_MF', 'Sale_CC', 'Sale_CL', 'Revenue_MF', 'Revenue_CC', 'Revenue_CL'], axis=1)

print(new_dataframe.shape)
new_dataframe.head()
```

```python
# preprocessing of data

from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.ensemble import ExtraTreesClassifier

new_dataframe.fillna(0, inplace=True)
new_dataframe[new_dataframe < 0] = 0
new_dataframe.head()
```

```python
#  spliting the data using "train_test_split" library of sci-kit learn

from sklearn.model_selection import train_test_split

X = new_dataframe.iloc[:, 1:29]
y = new_dataframe.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,)

# defining a model as "RandomForestRegressor"

model = RandomForestRegressor()
model.fit(X_train, y_train)

y_pred = model.predict(X_test) 

# calculating error between actual values and predicted values.

from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))


df = pd.DataFrame(X_test)
df['actual_revenue'] = y_test
df['predicted_revenue'] = y_pred
print('')

df.head()
```

```python
import matplotlib.pyplot as plt
_, ax = plt.subplots()

ax.scatter(x = range(0, y_test.size), y=y_test, c = 'blue', label = 'Actual', alpha = 0.3)
ax.scatter(x = range(0, y_pred.size), y=y_pred, c = 'red', label = 'Predicted', alpha = 0.3)

plt.title('Actual and predicted values')
plt.xlabel('Observations')
plt.ylabel('predicted')
plt.legend()
plt.show()
```
